# Importando as Bibliotecas

```python
# Biblioteca utilizada para carregar os dados
import pandas as pd
import basedosdados as bd
```

# Carregando os Dados

```python
tabela_avaliacao= bd.read_table(dataset_id='br_inep_saeb',
                   table_id='brasil',
                   billing_project_id="api-ads-388816")
```

# Visualizando a Tabela

```python
tabela_avaliacao.head()
```

## Valores que podem ser pesquisados

* ##  Anos disponivéis:

```python
# Salvando os valores cosiderados na tabela em uma lista
anos = list(tabela_avaliacao['ano'].unique())

# Ordenando a lista 
anos.sort()

# Devolvendo os valores da lista na tela
for n in anos:    # O loop for foi usado apenas para melhorar a visualização dos valores
    print(n)
```

* ##  Disciplinas disponivéis:

```python
# Salvando os valores cosiderados na tabela em uma lista
disciplinas = list(tabela_avaliacao['disciplina'].unique())

# Devolvendo os valores da lista na tela
for n in disciplinas:    # O loop for foi usado apenas para melhorar a visualização dos valores
    print(n)
```

# Criando a Pesquisa

```python
# Pedindo dados para o  usuário
ano = int(input("Informe o ano conforme opções acima: "))
disciplina = input("Informe a disciplinas conforme opções acima: ")
opcao = input("Informe com sim ou nao, se desejá baixar a busca: ")

# Gerando a busca, por filtragem
tabela_busca1 = tabela_avaliacao[tabela_avaliacao['ano'] == ano]

busca_final = tabela_busca1[tabela_busca1['disciplina'] == disciplina]

# Salvando ou não a pesquisa

if opcao in ['sim', 'Sim']:
    busca_final.to_excel('pesquisa_{}_{}.xlsx'.format(ano, disciplina))
    
elif opcao in ['nao', 'Nao', 'não', 'Não']:
    display(busca_final)

else:
    print("Digite apenas as opções informadas!")
    
```
